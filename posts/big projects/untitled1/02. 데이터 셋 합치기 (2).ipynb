```python
# 1) 데이터 가져오기
import pandas as pd
df2_p = pd.read_csv(r'C:\Users\brian\Desktop\JUNSOO\exhibition\data\StudentParent_counsel.csv' , encoding = 'cp949')
df3_c = pd.read_csv(r'C:\Users\brian\Desktop\JUNSOO\exhibition\data\SchoolCircle.csv' , encoding = 'cp949')
df4_t = pd.read_csv(r'C:\Users\brian\Desktop\JUNSOO\exhibition\data\SchoolTeacher.csv' , encoding = 'cp949')

## 결측치 row 삭제하기
df2_p['공시연도'].dropna(inplace = True)
df3_c['공시연도'].dropna(inplace = True)
df4_t['공시연도'].dropna(inplace = True)


# 2) 컬럼 선별하기
# 학교별 학부모 상담 건수
col_p = [ '상담실적(내부상담전문가)'        
          ,'상담실적(외부상담전문가)'
          ,'공시연도'
          ,'정보공시 학교코드']

# 학교별 동아리 활동
col_c = ['학생자율동아리 동아리수'        
         ,'창의적체험활동동아리 동아리수'
         ,'공시연도'
         ,'정보공시 학교코드']

# 학교별 교수학습공간
col_t = ['교수학습공간 일반교실'        
         ,'교수학습공간 교과교실'
         ,'교수학습공간 수준별교실'
         ,'교수학습공간 특별교실'
         ,'공시연도'
         ,'정보공시 학교코드']

```

# 학교별 학부모 상담건수
상담실적(내부상담전문가) - 학교 내부의 상담가능 전문가의 상담 건수
상담실적(외부상담전문가) - 학교 외부의 상담가능 전문가의 상담 건수

# 학교별 동아리 활동
학생자율동아리 동아리수 - 학생 자체적으로 만든 동아리수 
창의적체험활동동아리 동아리수 - 학교에서 시행중인 동아리수      #둘의 상관이 매우 낮아 둘다 포함

학생자율동아리 동아리 및 창의적체험활동동아리 동아리 예산지원현황 - 삭제 #결측치가 전체 데이터의 1/5

# 학교별 교수학습 공간
교수학습공간 일반교실 - 일반적인 교실의 수
교수학습공간 교과교실 - 특정 과목을 위해 마련된 교실의 수
교수학습공간 수준별교실 - 수준별수업을 위한 교실의 수
교수학습공간 특별교실 - 특정 교과에 대응한 실험, 실습등의 특별 학습 활동을 하기 위해 특정 설비가 갖춰진 교실의 수


```python
# 3) 데이터 프레임 작성

# 학교별 학부모 상담건수
counsel = df2_p[col_p]

# 학교별 동아리 활동
df3_c['공시연도'] = (pd.to_numeric(df3_c['공시연도'] , errors = 'coerce'))
df3_c = df3_c.dropna(subset = ['공시연도'])
df3_c['공시연도'] = df3_c['공시연도'].astype(int)

circle = df3_c[col_c]

# 학교별 교수학습공간
studyroom = df4_t[col_t]

# 4) key 값 생성

# 학교별 학부모 상담 건수
counsel = counsel.astype({'공시연도':'str'})
counsel['key'] = counsel['정보공시 학교코드'] + "-" + counsel['공시연도']

# 학교별 동아리 활동
circle = circle.astype({'공시연도':'str'})
circle['key'] = circle['정보공시 학교코드'] + "-" + circle['공시연도']

# 학교별 교수학습 공간
studyroom = studyroom.astype({'공시연도':'str'})
studyroom['key'] = studyroom['정보공시 학교코드'] + "-" + studyroom['공시연도']

# 5) key 값으로 인덱스 재 설정하기


(counsel
 .set_index(keys = 'key'
            ,inplace = True))

(circle
 .set_index(keys = 'key'
            ,inplace = True))

(studyroom
 .set_index(keys = 'key'
            ,inplace = True))

# 6) 필요없는 컬럼 지우기
## 필요없는 칼럼 지우기

(counsel.drop(columns = '공시연도' , inplace = True))
(counsel.drop(columns = '정보공시 학교코드' , inplace = True))

(circle.drop(columns = '공시연도' , inplace = True))
(circle.drop(columns = '정보공시 학교코드' , inplace = True))

(studyroom.drop(columns = '공시연도' , inplace = True))
(studyroom.drop(columns = '정보공시 학교코드' , inplace = True))

# 7) 데이터 합치기
final1 = pd.merge(counsel , studyroom, how = "outer" , on = 'key')
final2 = pd.merge(final1, circle, how = "outer", on = 'key')
final =  final2.drop_duplicates()


```

```python
# 8) 결측치 처리하기 
# 없는 교실
final['교수학습공간 교과교실'].fillna(0 , inplace = True)
final['교수학습공간 수준별교실'].fillna(0 , inplace = True)
final['교수학습공간 일반교실'].fillna(0 , inplace = True)
final['교수학습공간 특별교실'].fillna(0, inplace = True)

# 동아리수 평균
print(final['학생자율동아리 동아리수'].mean())    #8
final['학생자율동아리 동아리수'].fillna(8 , inplace = True)
print(final['창의적체험활동동아리 동아리수'].mean())   #34
final['창의적체험활동동아리 동아리수'].fillna(8 , inplace = True)



# 상담실적(내부 , 외부) 모두 52 ==> 삭제
final['상담실적(내부상담전문가)'].fillna(0 , inplace = True)
final['상담실적(외부상담전문가)'].fillna(0 , inplace = True)


```

```python
final.to_csv(r'C:\Users\brian\Desktop\JUNSOO\exhibition\data\dataset.csv' , encoding = "euc-kr")
```
