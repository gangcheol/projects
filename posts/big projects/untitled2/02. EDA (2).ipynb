```python
import pandas as pd 
df=pd.read_csv('결과.csv')
```

## 데이터 전처리

```python
# df의 ID_DATE라는 컬럼을 만들어서 ID컬럼값과 DATE컬럼 값을 합친 값으로 저장

df['ID_DATE'] = df['RESPOND_ID'].astype(str) + '_' + df['EXAMIN_BEGIN_DE'].astype(str)
```

```python
df.head()
```

```python
# df의 ID_DATE컬럼이 중복이 있으면 하나만 남기고 나머지는 삭제  
df.drop_duplicates(subset='ID_DATE', keep='first', inplace=True)
```

```python

df.rename(columns={'ID_DATE':'key'}, inplace=True)
df.to_csv('dataset3(key값 추가).csv', index=False)
```

```python
# df의 EXAMIN_BEGIN_DE 컬럼 순으로 정렬
df.sort_values(by='EXAMIN_BEGIN_DE', ascending=True)
```

```python
# df의 EXAMIN_BEGIN_DE 컬럼의 20211103 같은 데이ㅌ를 2021-11-03으로 변환
df['EXAMIN_BEGIN_DE'] = pd.to_datetime(df['EXAMIN_BEGIN_DE'], format='%Y%m%d')
```
