# 1. 데이터 로드 및 전처리

## (1) 데이터 로드

```python
train_df = pd.read_csv('QnA_train_data.csv')
train_df.head()
```

`-` 불필요한 열 삭제

```python
train_df.drop(["트랙", "지역"], axis = 1, inplace = True)
train_df.head()
```

`-` 개행문자(`"\n"`) 삭제

```python
train_df["문의내용"] = [i.replace("\n", " ") for i in train_df["문의내용"]]
```

## (2) 특수문자 제거

```python
removal_list =  "‘, ’, ◇, ‘, ”,  ’, ', ·, \“, ·, △, ●,  , ■, (, ), \", >>, `, /, #, ∼, =,ㆍ<,>, .,?, !,【,】, …, ◆,%, ₩"
def remove_special(sentence: str = None):

    sentence = re.sub("[.,\'\"’‘”“!?]", "", sentence)
    sentence = re.sub("[^ㄱ-ㅎ가-힣a-zA-Z\\s]", " ", sentence)
    sentence = re.sub("\s+", " ", sentence)
    sentence = sentence.translate(str.maketrans(removal_list, ' '*len(removal_list)))
    sentence = sentence.strip()
    sentence = sentence.replace('\n', ' ')

    return sentence
```

```python
train_df["특수문자제거"] = [remove_special(i) for i in train_df["문의내용"]]
train_df["특수문자제거"][:3]
```

## (3) 단어 분리

```python
train_df["단어분리"]= [i.split() for i in train_df["특수문자제거"]]
print(train_df["단어분리"][:2])
```

## (4) 조사, 인사말, 불용어 제거 함수 실행

```python
def remove_stopword(sent):
    stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','을','으로','자','에','와','한','이', '로', '에서', '하는', '하면', '하고', '요', '혹시', '합니다', '감사합니다', '안녕하세요']
    removed = [word for word in sent if not word in stopwords] # 불용어 제거
    return removed
```

```python
train_df["불용어제거"] = [remove_stopword(i) for i in train_df["단어분리"]]
train_df["불용어제거"].head()
```

## (5) 한글자 단어 제거

```python
train_df["한글자제거"] = [[j for j in i if len(j) >=2 ] for i in train_df["불용어제거"]]
train_df["한글자제거"][:5]
```

## (6) 문장 길이 측정

```python
cleansing_length =[len(i) for i in train_df["한글자제거"]]
train_df["cleansing_length"] = cleansing_length
```

## (7) 문의별 단어 수 시각화

```python
train_df.plot(x = "cleansing_length", kind = "hist",
                      backend  = "plotly", title = "단어길이 분포", 
                      height=  500, marginal = "box", opacity = 0.6)
```

## (8) padding 값 설정

`-` 자연어 처리에 입력의 길이를 똑같이 맞추기 위해 데이터의 앞 또는 뒤에 0을 삽입하는 Padding 기법을 적용

`-` 적절한 Padding 값을 도출하기 위해 max_length 변수를 생성하고 cleansing_length 변수의 기초통계량 90% 백분위수 값을 할당

```python
train_df.cleansing_length.describe()
```

`-` 자연어 처리에 입력의 길이를 똑같이 맞추기 위해 데이터의 앞 또는 뒤에 0을 삽입하는 Padding 기법을 적용

```python
max_length = train_df.cleansing_length.quantile(.9)
max_length
```

# 3. 단어사전 만들기

## (1) 단어 모으기

```python
word_list  = [j for i in train_df["한글자제거"] for j in i]
word_list[:5]
```

## (2) 단어 빈도수 카운트 

```python
word_count = Counter(word_list)
```

## (3) 전체 단어 빈도수 및 합계 확인

```python
word_frequency = word_count.values()
total_word = len(set(word_list))
total_frequency_sum = sum(word_count.values())

total_word, total_frequency_sum
```

## (4) 희귀 단어 빈도수 및 합계 확인

```python
rare_frequency = { i : j  for i,j in word_count.items() if j ==1}
rare_count = len(rare_frequency)
rare_frequency_sum = sum(rare_frequency.values())

rare_count, rare_frequency_sum
```

```python
# | code-fold : true
rare_count_percent = rare_count/total_word
rare_frequency_percent = rare_frequency_sum/total_frequency_sum


print("클렌징 데이터의 전체 단어 수: ", total_word)
print("데이터 내 전체 희귀 단어수: ",  rare_count)
print("\n")
print("데이터의 전체 단어의 빈도 수 합: ",  total_frequency_sum )
print("데이터 내 전체 희귀 단어의 빈도 수 합: ",   rare_frequency_sum)
print("전체 단어에서 희귀 단어가 차지하는 비율: ",  rare_frequency_percent)
```

## (5) 단어사전 크기 설정

`-` 단어사전크기 = 전체 단어수 - 희귀 단어수

```python
vocab_size = total_word-rare_count
vocab_size
```

## (6) 단어사전(인덱스화) 생성

```python
tokenizer = Tokenizer(num_words = vocab_size)
tokenizer.fit_on_texts(train_df["한글자제거"])

train_df["토큰화"] = tokenizer.texts_to_sequences(train_df["한글자제거"])
#train_df.head()
```

***

# 4. 모델 학습 데이터 생성

## (1) 문의유형 데이터 변경

`-` `data['문의유형']` 값 중 '코드1', '코드2'는 1로, 그 외 값들은 0으로 변경

```python
type_dict = {
    '코드1': 1,
    '코드2': 1,
    '웹': 0,
    '이론': 0,
    '시스템 운영': 0,
    '원격': 0
}

train_df["문의유형"] = train_df["문의유형"].map(type_dict)
```

## (2) `Y(target)` 데이터 분리

```python
y_data =  train_df["문의유형"]
```

## (3) X(Feature) 데이터 분리 및 padding 적용

```python
x_data = tf.keras.utils.pad_sequences(train_df["토큰화"], maxlen = int(max_length))
#x_data
```

## (4) 학습 데이터 분리

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size = 0.3, random_state = 2023)
```

***

## (5) 모델 생성

```python
import tensorflow as tf
from keras.layers import *
from keras.models import *
from keras.callbacks import *
from keras.backend import *
from keras.optimizers import *
```

### a. simple RNN 설계

```python
model = Sequential()
model.add(Embedding(input_dim = vocab_size + 1, output_dim = 128, input_length = int(max_length)))
## input_dim  : 단어 사전의 크기
## output_dim : 임베딩 차원
## input_length : 인풋 시퀀스의 길이

model.add(SimpleRNN(32))
model.add(Dense(32, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))

model.compile(loss = tf.keras.losses.binary_crossentropy,
                optimizer  = tf.keras.optimizers.Adam(0.01) )

model.summary()
```

####  1. 모델 학습

```python
from tensorflow.keras.callbacks import EarlyStopping
es = EarlyStopping(monitor = "val_loss",  ## 과적합 방지 기준
                                   min_delta = 0, ## 설정한 값으로 변화해야 모델 성능이 개선되었다고 판단하는 기준 ( Threshole)
                                    patience = 5, ## 성능이 개선되지 않을 떄 얼마나 참을건지
                                    verbose =1,
                                    restore_best_weights = True  ## 조기 종료 적용 후 최적 가중치를 모델의 전달
                                )

model.fit(x_train, y_train, epochs = 1000, validation_split = 0.2,
                batch_size = 32, callbacks = [es])
```

#### 2. 결과 예측

```python
predicted = model.predict(x_val)
result = [1 if np.round(i,1) >=0.5 else 0  for i in predicted ]
```

```python
print(confusion_matrix(y_val, result))
print("\n")
print(classification_report(y_val, result))
```

### b. LSTM 모델 설계

```python
clear_session()

model1 = Sequential()
model1.add(Embedding(input_dim = vocab_size + 1, output_dim = 128, input_length = int(max_length)))
model1.add(LSTM(32))
model1.add(Dense(32, activation = "relu"))
model1.add(Dense(1, activation = "sigmoid"))

model1.compile(loss = tf.keras.losses.binary_crossentropy,
                optimizer  = tf.keras.optimizers.Adam(0.01) )

model1.summary()
```

#### 1. 모델 학습

```python
history1 = model1.fit(x_train, y_train, epochs = 10, validation_split = 0.2,
                batch_size = 32, verbose = 0).history
```

#### 2. 결과 예측

```python
plt.plot(history1["loss"], label = "train_loss")
plt.plot(history1["val_loss"],label = "val_loss")
plt.legend()

predicted1 = model1.predict(x_val)
result1 = [1 if np.round(i,1) >=0.5 else 0  for i in predicted1]

print(confusion_matrix(y_val, result1))
print("\n")
print(classification_report(y_val, result1))
```

### c. LSTM 모델 설계 2(`+BatchNormalization`)

```python
clear_session()

model2 = Sequential()
model2.add(Embedding(input_dim = vocab_size + 1, output_dim = 128, input_length = int(max_length)))
model2.add(LSTM(32))
model2.add(BatchNormalization())
model2.add(Dense(32, activation = "relu"))
model2.add(BatchNormalization())
model2.add(Dense(1, activation = "sigmoid"))

model2.compile(loss = tf.keras.losses.binary_crossentropy,
                optimizer  = tf.keras.optimizers.Adam(0.01) )

model2.summary()
```

#### 1. 모델 학습


```python
history2 = model2.fit(x_train, y_train, epochs = 10, validation_split = 0.2,
                batch_size = 32, verbose = 0).history
```

```python
plt.plot(history2["loss"], label = "train_loss")
plt.plot(history2["val_loss"],label = "val_loss")
plt.legend()
```

#### 2. 결과 예측

```python
predicted2 = model2.predict(x_val)
result2 = [1 if np.round(i,1) >=0.5 else 0  for i in predicted2]

print(confusion_matrix(y_val, result2))
print("\n")
print(classification_report(y_val, result2))
```

### d. BiLSTM 모델 설계

```python
clear_session()

model3 = Sequential()
model3.add(Embedding(input_dim = vocab_size + 1, output_dim = 128, input_length = int(max_length)))
model3.add(Dense(1024, activation = "swish"))
model3.add(LSTM(32, return_sequences = True))
model3.add(BatchNormalization())
model3.add(GlobalAveragePooling1D())
model3.add(Dense(32, activation = "swish"))
model3.add(BatchNormalization())
model3.add(Dense(1, activation = "sigmoid"))

model3.compile(loss = tf.keras.losses.binary_crossentropy,
                optimizer  = tf.keras.optimizers.Adam(0.01) )

model3.summary()
```

#### 1. 모델 학습

```python
history3 = model3.fit(x_train, y_train, epochs = 10, validation_split = 0.2,
                                      batch_size = 32, verbose = 1).history
```

```python
plt.plot(history3["loss"], label = "train_loss")
plt.plot(history3["val_loss"],label = "val_loss")
plt.legend()
```

#### 2. 결과 예측

```python
predicted3 = model3.predict(x_val)
result3 = [1 if np.round(i,1) >=0.5 else 0  for i in predicted3]

print(confusion_matrix(y_val, result3))
print("\n")
print(classification_report(y_val, result3))
```
